Task 4: Scores by School Spending, Size, and Type

```python
import pandas as pd
```

```python
# Make a reference to the csv file path
filepath = "Output/hw4_schools_complete.csv"
# Import the csv file as a DataFrame
schools_complete = pd.read_csv(filepath, encoding = "utf-8")
schools_complete.head()
```

```python
# Rename the columns
schools_complete = schools_complete.rename(columns={"reading_score":"Average Reading Score", "math_score":"Average Math Score"})
schools_complete.head()
```

```python
# Classify schools into 4 groups by "budget per student", add a column to store this information
bins_bps = [0,585,615,645,675]
group_names = ["<$585","$585-615","$615-645","$645-675"]
schools_complete["Spending Ranges (Per Student)"] = pd.cut(schools_complete["Per_Student_Budget"], bins_bps, labels = group_names)
schools_complete.head()
```

```python
# Get the mean of each group
spending_group = schools_complete.groupby("Spending Ranges (Per Student)").mean()
spending_group.head()
```

```python
# Remove the columns we don't need
spending_group= spending_group.loc[:,["Average Math Score","Average Reading Score","% passing math","% passing reading","% Overall Passing Rate"]]
spending_group.head()
```

```python
# Classify schools into 3 groups by "school size", add a column to store this information
bins_size = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
schools_complete["School Size"] = pd.cut(schools_complete["size"], bins_size, labels = group_names)
schools_complete.head()
```

```python
# Get the mean of each group
size_group = schools_complete.groupby("School Size").mean()
size_group.head()
```

```python
# Remove the columns we don't need
size_group= size_group.loc[:,["Average Math Score","Average Reading Score","% passing math","% passing reading","% Overall Passing Rate"]]
size_group.head()
```

```python
# Group schools by "type",and get the mean of each type
type_group = schools_complete.groupby("type").mean()
type_group.head()
```

```python
# Remove the columns we don't need
type_group= type_group.loc[:,["Average Math Score","Average Reading Score","% passing math","% passing reading","% Overall Passing Rate"]]
type_group.head()
```

```python
# Save the results as excel files
spending_group.to_excel("Output/Task4_Scores_by_School_Spending.xlsx", encoding = "utf-8", index = False, header = True)
size_group.to_excel("Output/Task4_Scores_by_School_Size.xlsx", encoding = "utf-8", index = False, header = True)
type_group.to_excel("Output/Task4_Scores_by_School_Type.xlsx", encoding = "utf-8", index = False, header = True)
```
