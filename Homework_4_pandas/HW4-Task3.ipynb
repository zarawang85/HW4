Task 3: Average score by grade

```python
import pandas as pd
import numpy as np
```

```python
# Make a reference to the csv file path
filepath = "Resources/schools_complete.csv"
# Import the csv file as a DataFrame
schools_data = pd.read_csv(filepath, encoding = "utf-8")
schools_data.head()
```

```python
# Make a reference to the csv file path
filepath = "Resources/students_complete.csv"
# Import the csv file as a DataFrame
students_data = pd.read_csv(filepath, encoding = "utf-8")
students_data.head()
```

```python
# Collect only students from grade 9
# Get the average reading score of each school
# Get the average math score of each school
grade_9 = students_data.loc[students_data["grade"] == "9th"]
read_9th = grade_9.groupby("school_name").mean().loc[:,["reading_score"]]
math_9th = grade_9.groupby("school_name").mean().loc[:,["math_score"]]
```

```python
# Collect only students from grade 10
# Get the average reading score of each school
# Get the average math score of each school
grade_10 = students_data.loc[students_data["grade"] == "10th"]
read_10th = grade_10.groupby("school_name").mean().loc[:,["reading_score"]]
math_10th = grade_10.groupby("school_name").mean().loc[:,["math_score"]]
```

```python
# Collect only students from grade 11
# Get the average reading score of each school
# Get the average math score of each school
grade_11 = students_data.loc[students_data["grade"] == "11th"]
read_11th = grade_11.groupby("school_name").mean().loc[:,["reading_score"]]
math_11th = grade_11.groupby("school_name").mean().loc[:,["math_score"]]
```

```python
# Collect only students from grade 12
# Get the average reading score of each school
# Get the average math score of each school
grade_12 = students_data.loc[students_data["grade"] == "12th"]
read_12th = grade_12.groupby("school_name").mean().loc[:,["reading_score"]]
math_12th = grade_12.groupby("school_name").mean().loc[:,["math_score"]]
```

```python
# Combine reading average of all grades together
reading = pd.merge(read_9th,read_10th,left_index = True,right_index = True,suffixes=("_9th", "_10th"))
reading = pd.merge(reading,read_11th,left_index = True,right_index = True)
reading = pd.merge(reading,read_12th,left_index = True,right_index = True)
reading.head()
```

```python
# Rename the columns
reading = reading.rename(columns={'reading_score_9th':'9th', 'reading_score_10th':'10th', 'reading_score_x':'11th',
       'reading_score_y':'12th'})
reading.head()
```

```python
# Combine math average of all grades together
math = pd.merge(math_9th,math_10th,left_index = True,right_index = True,suffixes=("_9th", "_10th"))
math = pd.merge(math,math_11th,left_index = True,right_index = True)
math = pd.merge(math,math_12th,left_index = True,right_index = True)
math.head()
```

```python
# Rename the columns
math = math.rename(columns={'math_score_9th':'9th', 'math_score_10th':'10th', 'math_score_x':'11th',
       'math_score_y':'12th'})
math.head()
```

```python
# Save the results as excel files
reading.to_excel("Output/Task3_Reading_Average_by_grade.xlsx", encoding = "utf-8", index = False, header = True)
math.to_excel("Output/Task3_Math_Average_by_grade.xlsx", encoding = "utf-8", index = False, header = True)
```
